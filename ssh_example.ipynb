### Generating "random phase" samples of 2D field: time-mean SSH
---

```python
import matplotlib.pyplot as plt
import scipy.io as sio
import numpy as np

from ranthymetry import gen_samples

%matplotlib inline

# load data
data = sio.loadmat('timeMeanSSH.mat')
ssh = data['timeMeanSSH']

# original ssh map
fig, ax = plt.subplots( 1, 1, figsize=(13,5) )

im = ax.imshow( ssh, origin='lower', cmap='inferno' )
cbar = plt.colorbar( im, ax=ax )

ax.set_title('Time-mean satellite SSH in North Atlantic')
ax.set_xlabel('Lon')
ax.set_ylabel('Lat')
ax.set_xticks( np.linspace(0,280,8) )
ax.set_xticklabels( ['80W','70W','60W','50W','40W','30W','20W','10W'] )
ax.set_yticks( [20,60,100] )
ax.set_yticklabels( [ '35N','45N','55N'] )

plt.show()
```

```python
# select sub-region with no coast
f = ssh[20:100,120:240]

print f.shape

# plot field we are going to generate
# random samples of
plt.imshow(f, origin='lower', cmap='inferno' )
```

```python
# generate 10 random samples with
# sample 2D power spectrum
f_samples = gen_samples( f, n_samples=10 )

for n in range(3) :

    plt.imshow( f_samples[n,:,:], origin='lower', cmap='inferno' )
    plt.show()
```
